```python
def get_balance_sheet(file):
    dict_df = pd.read_excel(file, sheet_name=['СБД','ОДТ'])
    balance_df = dict_df.get('СБД')
    return balance_df
```

```python
balance_df = get_balance_sheet(file)
```

```python
def get_income_sheet(file):
    dict_df = pd.read_excel(file, sheet_name=['СБД','ОДТ'])
    income_df = dict_df.get('ОДТ')
    return income_df
```

```python
income_df = get_income_sheet(file)
```
